```python
import numpy as np
import matplotlib.pyplot as plt

eniac = np.load('performance_eniac.npy')
eniac_width = np.load('performance_eniac_width.npy')
ospoe_width = np.load('performance_ospoe_width.npy')
ospoe_width_decay = np.load('performance_ospoe_width_decay.npy')
```

```python
# plt.plot(eniac.mean(axis=1), label='eniac')
# plt.fill_between(list(range(len(eniac))), eniac.mean(axis=1)-eniac.std(axis=1), eniac.mean(axis=1)+eniac.std(axis=1), alpha=.1)

plt.plot(eniac_width.mean(axis=1), label='eniac width')
plt.fill_between(list(range(len(eniac))), eniac_width.mean(axis=1)-eniac_width.std(axis=1), eniac_width.mean(axis=1)+eniac_width.std(axis=1), alpha=.3)

plt.plot(ospoe_width.mean(axis=1), label='ospoe width')
plt.fill_between(list(range(len(eniac))), ospoe_width.mean(axis=1)-ospoe_width.std(axis=1), ospoe_width.mean(axis=1)+ospoe_width.std(axis=1), alpha=.5)

# plt.plot(ospoe_width_decay.mean(axis=1), label='ospoe width decay')
# plt.fill_between(list(range(len(eniac))), ospoe_width_decay.mean(axis=1)-ospoe_width_decay.std(axis=1), ospoe_width_decay.mean(axis=1)+ospoe_width_decay.std(axis=1), alpha=.1)

plt.legend()
plt.show()
```
